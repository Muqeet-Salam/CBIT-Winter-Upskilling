````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score

columns = [
    "class", "alcohol", "malic_acid", "ash", "alcalinity_of_ash", "magnesium",
    "total_phenols", "flavanoids", "nonflavanoid_phenols", "proanthocyanins",
    "color_intensity", "hue", "OD280_OD315_of_diluted_wines", "proline"
]

# Load the dataset
df = pd.read_csv("/content/wine.data", sep=",", header=None, names=columns)
````

````python
x = df.iloc[:, [1, 13]].values
y = df.iloc[:, [13]].values
````

````python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
````

````python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

````python
linear_model = LinearRegression()
linear_model.fit(X_train_scaled, y_train)
y_pred_linear = linear_model.predict(X_test_scaled)
````

````python
y[31]
````

````python
poly_features = PolynomialFeatures(degree=2)
X_train_poly = poly_features.fit_transform(X_train_scaled)
X_test_poly = poly_features.transform(X_test_scaled)
````

````python
non_linear_model = LinearRegression()
non_linear_model.fit(X_train_poly, y_train)
y_pred_non_linear = non_linear_model.predict(X_test_poly)
````

````python
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train_scaled, y_train)
y_pred_ridge = ridge_model.predict(X_test_scaled)
````

````python
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train_scaled, y_train)
y_pred_lasso = lasso_model.predict(X_test_scaled)
````

````python
print("MSE:", mean_squared_error(y_test, y_pred_linear))
print("R2 Score:", r2_score(y_test, y_pred_linear))
````

````python
print("\nNon-linear Regression:")
print("MSE:", mean_squared_error(y_test, y_pred_non_linear))
print("R2 Score:", r2_score(y_test, y_pred_non_linear))
````

````python
print("\nRidge Regression:")
print("MSE:", mean_squared_error(y_test, y_pred_ridge))
print("R2 Score:", r2_score(y_test, y_pred_ridge))
````

````python
print("\nLasso Regression:")
print("MSE:", mean_squared_error(y_test, y_pred_lasso))
print("R2 Score:", r2_score(y_test, y_pred_lasso))
````

Handling non linear classes

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
````

````python
np.random.seed(0)
feat1 = np.random.randn(200,2)
feat1
````

````python
target_xor = np.logical_xor(feat1[:,0] > 0, feat1[:,1] > 0)
target1 = target_xor.astype(int)
````

````python
plt.scatter(feat1[target1 == 0, 0], feat1[target1 == 0, 1], color="red", label="Class 0")
plt.scatter(feat1[target1 == 1, 0], feat1[target1 == 1, 1], color="blue", label="Class 1")
plt.legend()
plt.title("XOR Dataset")
plt.show()
````

````python
scaler = StandardScaler()
feat_scaled = scaler.fit_transform(feat1)
````

````python
svc = SVC(kernel="rbf", random_state=0, gamma=.10, C=10.0)
````

````python
modelnonlinear = svc.fit(feat_scaled,target1)
````

````python
modelnonlinear.predict([feat_scaled[0]])
````

````python
target1[0]
````

````python
x_min, x_max = feat_scaled[:, 0].min() - 1, feat_scaled[:, 0].max() + 1
y_min, y_max = feat_scaled[:, 1].min() - 1, feat_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))
Z = modelnonlinear.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdYlBu)
plt.scatter(feat_scaled[target1 == 0, 0], feat_scaled[target1 == 0, 1], color="red", label="Class 0", edgecolor="k")
plt.scatter(feat_scaled[target1 == 1, 0], feat_scaled[target1 == 1, 1], color="blue", label="Class 1", edgecolor="k")
plt.title("Non-Linear Decision Boundary (RBF Kernel)")
plt.xlabel("Feature 1 (Scaled)")
plt.ylabel("Feature 2 (Scaled)")
plt.legend()
plt.show()
````
